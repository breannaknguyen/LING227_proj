## Building a classifier

```python
class SentimentClassifier(nn.Module):

  def __init__(self):
    super(SentimentClassifier, self).__init__()

    # Define the tokenizer
    self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    # Define the model
    self.bert = BertModel.from_pretrained("bert-base-uncased")

    self.output = nn.Linear(768, 1)

    # Define the sigmoid function
    self.sigmoid = nn.Sigmoid()

  def forward(self, inp):

    # Tokenize the input
    encoded_input = self.tokenizer(inp, return_tensors='pt')
    # Here we define the process of proceeding through the network:
    for key in encoded_input:
      encoded_input[key] = encoded_input[key].to(device=device)

    # Get the BERT hidden state over [CLS]
    emb = self.bert(**encoded_input)["last_hidden_state"][0][0]

    emb = emb.detach()

    # Pass through our linear output layer and sigmoid to
    # get the prediction
    outp = self.sigmoid(self.output(emb))

    return outp

  def predict(self, sentence):
    outp = self.forward(sentence)

    if outp.item() >= 0.50:
      return "positive", outp
    else:
      return "negative", outp
```

## Training functions
To train the network, we will be able to exactly copy our training functions from the older Colab notebook where we introduced Pytorch. Refer to that notebook for more details about these functions.

```python
loss_function = nn.BCELoss()
def compute_validation_loss_sen(model):

  total_loss = 0
  total_correct = 0
  count_examples = 0

  # Loop over all examples in the validation set
  for example in sst2["validation"]:

    # Get the sentence and the label for this example
    sentence = example["sentence"]
    label = torch.FloatTensor([example["label"]]).to(device=device)

    # Get the model's predicted output
    predicted = model(sentence)

    # Compute the loss for this prediction
    loss = loss_function(predicted, label)

    # Update our running totals for the number of
    # examples, the total loss, and the number correct
    total_loss += loss
    count_examples += 1

    if predicted.item() >= 0.5 and example["label"] == 1:
      total_correct += 1
    elif predicted.item() < 0.5 and example["label"] == 0:
      total_correct += 1

  # Return average loss and average accuracy
  return total_loss / count_examples, total_correct / count_examples
```

```python
loss_function = nn.BCELoss()
def compute_validation_loss_eth(model):

  total_loss = 0
  total_correct = 0
  count_examples = 0

  # Loop over all examples in the validation set
  for example in ethics["validation"]:

    # Get the sentence and the label for this example
    input = example["input"]
    label = torch.FloatTensor([example["label"]]).to(device=device)

    # Get the model's predicted output
    predicted = model(input)

    # Compute the loss for this prediction
    loss = loss_function(predicted, label)

    # Update our running totals for the number of
    # examples, the total loss, and the number correct
    total_loss += loss
    count_examples += 1

    if predicted.item() >= 0.5 and example["label"] == 1:
      total_correct += 1
    elif predicted.item() < 0.5 and example["label"] == 0:
      total_correct += 1

  # Return average loss and average accuracy
  return total_loss / count_examples, total_correct / count_examples
```

```python
# First instantiate the model.
# Note that we must move the model onto our GPU ("device")
cls = SentimentClassifier().to(device=device)

# For our optimizer, we will use Adam with a
# learning rate of 0.001
optimizer = torch.optim.Adam(cls.parameters(), lr=0.001)

# Now we loop over the training set and update the weights
# for each example!
loss = 0
for index, example in enumerate(sst2["train"]):

  # Get the sentence and label for this example
  sentence = example["sentence"]
  label = torch.FloatTensor([example["label"]]).to(device=device)

  # Get the model's prediction
  predicted = cls(sentence)

  # Compute the loss
  loss += loss_function(predicted, label)

  # We will only update the weights after every 64 examples.
  # This is called using a "batch".
  # This is actually an inefficient way to implement it; there
  # are better ways to do batches. But this approach is easiest
  # to implement.
  if index % 64 == 0:
    # Backpropagate, then update the weights using optimizer.step, and
    # then zero out the gradients since we have now used them
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    loss = 0

  # We will evaluate performance on the validation set after every
  # 1000 training examples. Hopefully the loss will decrease and
  # the accuracy will increase.
  if index % 1000 == 0:
    loss, accuracy = compute_validation_loss_eth(cls)
    print(index, loss.item(), accuracy)
```
